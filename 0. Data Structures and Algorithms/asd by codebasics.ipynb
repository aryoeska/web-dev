## BIG O NOTATION

is used to measure how running time or space requirements for your program grow as input size grows

*digunakan untuk mengukur waktu atau memori yang dibutuhkan untuk program bertambah sesuai dengan bertambahnya ukuran input*




you can see, in **array** as the size grows/increases the time also grows/increases. It's linearly proportional between both of them

### *time = a x n +b*

*a* is constant representing the time required per unit operation, such as (1,2,3,4,5,6,...)

*n* is the input size

*b* is the fixed time required to start the function like (20ms, 10ms, etc)

**this is the *mathematical representations* of BIG O NOTATION : *O(n)***

### The BIG O NOTATION HAS some rules to apply :

1. keep the fastest growing term ==> *time = a x b* 
2. drop the constast ==> *time = O(n)*

### Question 1
Build the squaring numbers function by taking the list

```python
def get_squared_numbers(numbers):
    squared_numbers = []
    for n in numbers :
        squared_numbers.append(n*n)
    return squared_numbers

a = [6,7,8,9]
get_squared_numbers(a)
```

### Example without using big O notation
We want build the duplicate and unique numbers function.

#### Without Big O Notation principle

```python
numbers = [3, 6, 7, 1, 6, 2, 8, 9, 3]

unique_number = []
duplicate_number = []

for i in range(len(numbers)):
    is_duplicate = False
    for j in range(len(numbers)):
        if i != j and numbers[i] == numbers[j]:
            is_duplicate = True
            break
    if is_duplicate:
        if numbers[i] not in duplicate_number:
            duplicate_number.append(numbers[i])
            print(numbers[i], 'is duplicated')
    else:
        if numbers[i] not in unique_number:
            unique_number.append(numbers[i])
            print(numbers[i], 'is unique')
```

#### With Big O Notation Principle

```python
frequency = {}

for i in numbers:
    if i in frequency:
        frequency[i] += 1
    else:
        frequency[i]=1

for i in numbers:
    if frequency[i] > 1:
        print(i, 'is duplicate')
    else :
        print(i, 'is unique')
```

```python
print(frequency)
```

```python
from collections import Counter

count = Counter(numbers)

for i in numbers:
    if count[i] > 1:
        print(f'{i} is duplicate')
    if count[i] == 1:
        print(f'{i} is unique')
```

### Question 2
Build the function of price to earnings (P/E). Calculate the stock price valuation by the stock price devided by earning per share for each stock

```python
def find_first_pe(price, eps, i):
    pe = price[i]/eps[i]
    return pe

price_list = [200000, 80000, 900000, 400000]
eps_list = [3, 5, 7, 6]
index = 3 # stock 4, which mean 400000/6

find_first_pe(price_list, eps_list, index)
```
